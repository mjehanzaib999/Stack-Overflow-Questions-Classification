```python
# Extract only title, markdown, and the open status from the data for training process
df = df[['Title', 'BodyMarkdown', 'OpenStatus']]
df.shape
df
```

```python
# Convert the OpenStatus labels into numbers for supervised training

df['OpenStatus'] = df['OpenStatus'].map({'open': 0, 'not a real question': 1, 'off topic': 2, 'not constructive': 3, 'too localized': 4})
df
```

```python
# Data preprocessing for model training
from sklearn.model_selection import StratifiedShuffleSplit

train_df = df['Title']+ ''+ df['BodyMarkdown']
train_df = train_df.dropna()
labels_df = df['OpenStatus'].dropna().astype('int')
# Applying the method
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=0)
# sss.get_n_splits(train_df, labels_df)
train_idx = []
test_idx = []
print(sss.split(train_df, labels_df))
for i, (train_index, test_index) in enumerate(sss.split(train_df, labels_df)):
  print(f"Fold {i}:")
  print(f"  Train: index={train_index}")
  train_idx.append(train_index)
  test_idx.append(test_index)
  print(f"  Test:  index={test_index}")

print(len(train_idx[0]))
print(len(test_idx[0]))

```

```python
#Check the stratified folds for train
from collections import Counter

label_check_lst = []
for idx in train_idx[0]:
  label_check_lst.append(labels_df[idx])

Counter(label_check_lst)
```

```python
# Check the strarified folds for test
label_check_lst = []
for idx in test_idx[0]:
  label_check_lst.append(labels_df[idx])
  
Counter(label_check_lst)
```

```python
# Separate the folds of train and test 

X_train = [train_df[idx] for idx in train_idx[0]]
X_test = [train_df[idx] for idx in test_idx[0]]
Y_train = [labels_df[idx] for idx in train_idx[0]]
Y_test = [labels_df[idx] for idx in test_idx[0]]

print(len(X_train))
print(len(X_test))

print(len(Y_train))
print(len(Y_test))
```

```python
# Data preprocessing
import re
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk import pos_tag

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger') 

def text_processing(text):
  """Function for performing basic text processing
  Args:
    text: Text from each stack overflow question"""
  #text = X_train[0].lower()
  # print(re.findall(r"[0-9]+", text))
  processed_text = re.sub('[\[{()}\]0-9+]'," ",text) # Handling the brackets and digits
  processed_text = processed_text.replace("  "," ") #  Handling extra whitespaces
  processed_text = re.sub('[\W\s]+'," ", processed_text).replace("  ", " ") # Handling punctuation
  # print(processed_text)

  # Stopwords
  stop_words = stopwords.words('english')

  #Tokenize
  words = word_tokenize(processed_text)
  # Filter the words in the tokens that are not stop words 
  filtered_words = [word for word in words if word not in stop_words]

  # Stemming
  porter = PorterStemmer()
  stemmed = [porter.stem(word) for word in filtered_words] # Stem the words that are not stop words
  #print(stemmed)

  # POS Tags for each token
  pos = pos_tag(filtered_words)
  # print(pos)

  # Convert the stemmed words into a string
  stemmed_string =' '.join(stemmed)
  return stemmed_string
```

```python
for index in range(len(X_train)):
  X_train[index] = text_processing(X_train[index])

for index in range(len(X_test)):
  X_test[index] = text_processing(X_test[index])
```

```python
# Let's covvert words to numbers using TF-IDF 
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features = 10000)  # it contains only 10k features (fixed!)

xtrain_tfidf = vectorizer.fit_transform(X_train).toarray()  # converting words to numbers for train data 
xtest_tfidf = vectorizer.transform(X_test).toarray()   
```

```python
# Naive Bayes 
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()

clf.fit(xtrain_tfidf, Y_train)

predicted_naive = clf.predict(xtest_tfidf)
```

```python
# Metrics :) 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix 

print('Accuracy Score \n',accuracy_score(predicted_naive, Y_test))
print('Confusion Matrix \n', confusion_matrix(predicted_naive, Y_test))
print('Classification Report \n', classification_report(predicted_naive, Y_test))
```
